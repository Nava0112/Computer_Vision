```python
from matplotlib.image import imread
import  matplotlib.pyplot as plt
import cv2
import numpy as np
```

CONVERT TO CMY:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = 'wallpaperflare.com_wallpaper (7).jpg'
image_bgr = cv2.imread(image_path)

image_rgb = image_bgr[:, :, ::-1]


plt.imshow(image_rgb)
plt.title("Original Image (RGB)")
plt.axis('off')
plt.show()
image_rgb_normalized = image_rgb / 255.0

cmy = 1 - image_rgb_normalized

cmy_scaled = (cmy * 255).astype(np.uint8)

plt.imshow(cmy_scaled)
plt.title("Converted Image (CMY)")
plt.axis('off')
plt.show()
```


CONVERT TO HSV:

```python
image_bgr_for_hsv = image_rgb[:, :, ::-1]
image_hsv = cv2.cvtColor(image_bgr_for_hsv, cv2.COLOR_RGB2HSV)
image_rgb = image_bgr[:, :, ::-1]

plt.imshow(image_hsv)
plt.title("Converted Image (HSV)")
plt.axis('off')
plt.show()
```

CONVERT TO ONLY RED CHANNEL:

```python
image_path = 'wallpaperflare.com_wallpaper (7).jpg'
image_bgr = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

red_channel = image_rgb.copy()
red_channel[:, :, 1] = 0
red_channel[:, :, 2] = 0

plt.imshow(red_channel)
plt.title('Red Channel')
plt.axis('off')
plt.show()
```
